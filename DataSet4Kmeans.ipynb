```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
datos = pd.read_csv('Datos4_1.csv')
x = datos.iloc[:, 0:2]
y = datos.iloc[:, 2]
```

```python
km = KMeans(n_clusters = 5)
km.fit(x)
```

```python
predicciones = km.predict(x)
score = metrics.adjusted_rand_score(y, predicciones)
print(score)
```

```python
#graficar datos y centroides
centroides = km.cluster_centers_
plt.scatter(x.iloc[:,0], x.iloc[:,1], c = y, s = 100)
plt.scatter(centroides[:, 0], centroides[:, 1], c = ['r', 'r', 'r', 'r', 'r'] ,marker = 'x', s = 150)
plt.grid()
plt.show()
```
